#### Creating Data Frames
DataFrames are 2-dimensional labeled data structures in Pandas. They can be thought of as a collection of Series objects that share the same index. We can create a DataFrame from a variety of data sources, including lists, dictionaries, and NumPy arrays.

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 34, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary' : [70000, 80000, 90000, 100000]
}
```

```python
df_dic = pd.DataFrame(data)
df_dic
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]

df_list = pd.DataFrame(data_list) # Creating DataFrame from list of lists without column names
print(df_list, "\n")

# Now giving names to columns while creating DataFrame
df_list_named = pd.DataFrame(data_list, columns=['Name', 'Age', 'City', 'Salary'])
print(df_list_named)
```

#### Selection and Indexing of Columns

```python
my_df = pd.DataFrame(data_list, columns=['Name', 'Age', 'City', 'Salary'])
```
