```python
import numpy as np
import pandas as pd
```

**1. Group By** is a powerful method in pandas that allows you to split your data into groups based on some criteria, apply a function to each group independently, and then combine the results back into a single DataFrame. This is particularly useful for performing operations like aggregation, transformation, and filtering on subsets of your data.

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
my_df = pd.DataFrame(data)
```

```python
my_df
```

```python
# Group by Category and calculate the sum of Sales
sales_sum = my_df.groupby('Category')['Sales'].sum()
print(sales_sum)
```

```python
# Group by Store and calculate the sum of Sales
sales_sum_storewise = my_df.groupby('Store')['Sales'].sum()
print(sales_sum_storewise)
```

```python
# Group by multiple columns
# Group by Category and Store

sum_of_sales = my_df.groupby(['Category', 'Store'])['Sales'].sum()
print(sum_of_sales)
```

```python
my_df['Sales'].mean()
```
