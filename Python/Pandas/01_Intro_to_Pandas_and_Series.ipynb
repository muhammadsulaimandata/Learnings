**Pandas Series** 
A series is one dimentional array capable of holding any data type. It is like a column in a table. It has an index and values. We can create a series using the `pd.Series()` function.

#### Creating Series
There are multiple ways to create a Series in Pandas. We can create a Series from a list, a NumPy array, a dictionary, or even from a scalar value.

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10, 20, 30]
my_array = np.array([10, 20, 30])
my_dict = {1: 10, 2: 20, 3: 30}
```

```python
print(my_list)
```

```python
pd.Series(data=my_list)
```

```python
pd.Series(data=my_list, index=labels)
```

```python
pd.Series(data=my_array, index=labels)
```
