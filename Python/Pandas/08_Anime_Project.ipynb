```python
import numpy as np
import pandas as pd
```

```python
# Loading Data
df = pd.read_csv(r'anime.csv')
```

```python
df.head()
```

```python
print(df.loc[0]) # for viewing only first row of the dataframe
print(f'\nOnly Tile, \n{df.loc[1]["Title"]}') # for viewing only the title of the first row of the dataframe
```

```python
# Making a Column for Episodes
def extract_episode(txt):
    data = ""
    Check = False
    for i in txt:
        if i == ")":
            break
        elif i == "(":
            Check = True
        elif Check == True:
            data += i
    return data

df['Episodes'] = df['Title'].apply(extract_episode)
df
```

```python
# Removing eps from episodes column

df['Episodes'] = df['Episodes'].str.replace(' eps', '')
df['Episodes'] = df['Episodes'].astype(int)
df.head()
```
