```python
import numpy as np
import pandas as pd
```

## Missing Data

**1. Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
my_df = pd.DataFrame(data)
```

```python
my_df
```

```python
print(my_df.isna(), "\n") # Check for missing values
print(my_df.isna().sum()) # Count missing values in each column
```

```python
my_df.isna().any() # Check if any missing values exist in the DataFrame
```

**2. Removing Missing Data**

```python
my_df.dropna() # Drop rows with missing values
```

```python
my_df.dropna(thresh = 3) # Drop rows that have less than 3 non-NA values
```

**3. Filling Missing Values**

```python
my_df.fillna(0) # Fill missing values with 0
```

```python
values = {'A': 100, 'B' : 200, 'C': 300, 'D': 400}
my_df.fillna(value = values) # Fill missing values with different values for each column
```

```python
my_df.fillna(my_df.mean())
```
