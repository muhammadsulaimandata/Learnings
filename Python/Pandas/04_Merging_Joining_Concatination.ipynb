```python
import numpy as np
import pandas as pd
```

**1. Merging two DataFrames:**

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
print("Employees DataFrame:")
print(employees)
print("\nSalaries DataFrame:")
print(salaries)
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'inner') # Inner join
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'left') # Left join
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'right') # Right join
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'outer') #  Outer join 
```

**2. Concatenation**

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
pd.concat([df1, df2]) # Concatenate along rows
```

```python
pd.concat([df1, df2], axis = 1) # Concatenate along columns
```

**3. Joining**

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1.join(df2) # Join on index
```

```python
df2.join(df1) # Join on index with df2 as the left DataFrame    
```
