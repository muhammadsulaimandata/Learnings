#### Array Operations

```python
import numpy as np
```

```python
# Adding two Arrays

# For doing this the array shape & size should be same
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
print(f"Sum of Array a1 {a1} and Array a2 {a2} is {a1 + a2}")
```

```python
# Subtracting two Arrays
a1 - a2
```

```python
# Multiplying two Arrays
a1 * a2
```

```python
# Dividing two Arrays
a1 / a2
```

```python
# Floor Division
a1 // a2
```
