#### Array Operations

```python
import numpy as np
```

```python
# Adding two Arrays

# For doing this the array shape & size should be same
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
print(f"Sum of Array a1 {a1} and Array a2 {a2} is {a1 + a2}")
```

```python
# Subtracting two Arrays
a1 - a2
```

```python
# Multiplying two Arrays
a1 * a2
```

```python
# Dividing two Arrays
a1 / a2
```

```python
# Floor Division
a1 // a2
```

```python
a1 ** a2
```

#### Broadcasting

```python
# Broadcasting is a powerful mechanism that allows NumPy to work with arrays of different shapes during 
# arithmetic operations. It automatically expands the smaller array to match the shape of the larger array, 
# enabling element-wise operations without the need for explicit loops.

my_list = [1,2,3,4,5]
my_array = np.array(my_list)
my_list + 10 # To add 10 to each element of the list we need to use list comprehension or a loop so instead of this we should use numpy array which supports broadcasting
my_array + 10   
```

**Deep and Shallow Copy** Shallow copy means to copy something and do some operations and the original is not affected and there is an another copy of original data which is stored in an another address

```python
my_array = np.arange(1,21)
print(my_array)
```

```python
arr_slice = my_array[:5]
print(f"Original Sliced Array:{arr_slice}\n")
arr_slice = arr_slice * 2
print(f"Array after Multiplication with 2:{arr_slice}\n")
```

#### Matrix Operations

```python
A = np.arange(1,5).reshape(2,2)
B = np.arange(5,9).reshape(2,2)
```

```python
print(f"Array A:\n{A}\n")
print(f"Array B:\n{B}\n")
```

```python
# Multiplication
A @ B
A * B # Element-wise multiplication and doesn't represent the dot product of A and B
print(f"Dot Product of A and B is:\n{A @ B}\n")
print(f"Element-wise Product of A and B is:\n{A * B}\n")
```

```python
# Another method of doing dot product is using np.dot() function
np.dot(A,B)
print(f"Dot Product of A and B using np.dot() function is:\n{np.dot(A,B)}\n")
```

```python
# Transpose
A.T
```

### Advance Array Manipulation

```python
# Splitting the array
c = np.arange(16).reshape(4,4)
print(f"Simple Array of 4 by 4:\n {c}\n")
```

```python
splited = np.hsplit(c,2) # c is the array and 2 are the parts in which we want to split this array
print(f"Array after spliting:\nSplit 1\n {splited[0]},\nSplit 2\n {splited[1]}\n")
```
