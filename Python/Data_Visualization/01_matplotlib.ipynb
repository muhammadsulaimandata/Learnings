```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**1. Basic Matplot Functions**

```python
x = np.linspace(0, 5, 11)
print(f'X Array: {x}')

y = x ** 2
print(f'\nY Array: {y}')
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My Plot Title')
plt.legend(['Y = X^2'])
```

**2. Subplots in Matplotlib**

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(y, x)
plt.xlabel('Y Label')
plt.ylabel('X Label')
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x**2, y)
plt.xlabel('X Square as X Label')
plt.ylabel('Y Label')
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot 4')

plt.tight_layout()
plt.show()
```
