### 1. Loading the Libraries and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('IPL.csv')
df.head()
```

##### 1. Which team won the most matches ?

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index, x = match_wins.values, palette= 'viridis')
plt.title('Most Matches Wining Teams')
```

#### 2. Check how many people choose to field who won the toss?

```python
sns.countplot(x = df['toss_decision'], palette= 'rainbow')
plt.title("Toss Decision Trend")
```

##### 3. Toss winner VS Match winner

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
print(f'Count of Match wins when also won toss is {count}.')
percentage = (count * 100) / df.shape[0]
print(f'Percentage of Match wins when tossed were win is {percentage.round(2)} percent.')
```

##### 6. Top Scorer

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending= False)
top_5 = high.head(5)
```

```python
sns.barplot(x = top_5.values, y = top_5.index, palette= 'rainbow')
```

#### 7. Highest Wicket Taker

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending = False).head(10)
```

```python
top_bowlers
```

```python
top_bowlers.plot(kind= 'barh')
plt.xlabel('Wickets')
plt.ylabel('Bowlers')
plt.title("Top Bowlers with Wickets")
```
