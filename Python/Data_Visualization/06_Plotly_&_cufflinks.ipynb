```python
# 1. Import ONLY the modern, working libraries
import seaborn as sns
import pandas as pd
import plotly.express as px  # <-- This is the modern replacement for cufflinks!

# 2. Load your data
tips = sns.load_dataset('tips')

# 3. Create and show the interactive plot 
# We use px.histogram instead of .iplot() for a single column of numbers
fig = px.histogram(tips, x='total_bill', title="Total Bill Distribution")
fig.show()
```
